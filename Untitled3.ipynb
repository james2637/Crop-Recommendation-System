```python
# ============================================================
# CELL 1 — INSTALL DEPENDENCIES
# Copy into a separate cell, run it, wait for the ✓ tick.
# ============================================================
!pip install -q gradio requests-cache retry-requests joblib scikit-learn pandas numpy matplotlib openpyxl

# ── Optional: mount Google Drive if your dataset lives there ──
# from google.colab import drive
# drive.mount('/content/drive')
# ─────────────────────────────────────────────────────────────

# ============================================================
# CELL 2 — MAIN APPLICATION CODE (v10.0 — Bulletproof Edition)
# ============================================================

import os, hashlib, warnings, io
import pandas as pd
import numpy as np
import requests_cache
import joblib
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import gradio as gr
from retry_requests import retry
from datetime import datetime, timedelta

# sklearn
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

warnings.filterwarnings("ignore")

# --- 1. CACHING & INFRASTRUCTURE ---
cache_session = requests_cache.CachedSession('/tmp/assam_cache', expire_after=3600)
retry_session = retry(cache_session, retries=5, backoff_factor=0.2)

print("=" * 72)
print("  🌾 ASSAM FARMER AI v10.0 — Bulletproof Colab Edition")
print("=" * 72)

# --- 2. CORE ENGINE CLASS ---
class AssamAIEngine:
    def __init__(self):
        self.is_ready = False
        self.data_loaded = False
        self.df = None
        self.rf_cat = None
        self.rf_pri = None
        self.preprocessor = None
        self.le_cat = None
        self.le_pri = None
        self.FEATURE_COLS = []
        self.CAT_COLS = []
        self.metrics = {}

        # Lookup lists
        self.DISTRICTS, self.CROPS, self.SOIL_TYPES = [], [], []
        self.WATER_SOURCES, self.FIELD_CONDITIONS, self.DRAINAGE_LEVELS = [], [], []
        self.PREVIOUS_CROPS, self.PEST_HISTORIES, self.FERTILIZERS = [], [], []
        self.PESTICIDES, self.IRRIGATION_SOURCES = [], []
        self.WATER_AVAIL_LEVELS, self.WATERLOGGING_FREQS = [], []
        self.BLOCKS_BY_DISTRICT, self.CROP_INFO, self.CATEGORY_TEXTS = {}, {}, {}
        self.DISTRICT_INFO = {}
        self.DISTRICT_SOIL_AVG = {}

        self.DISTRICT_COORDS = {
            "Baksa": (26.68, 91.56), "Barpeta": (26.32, 91.00),
            "Biswanath": (26.73, 93.15), "Bongaigaon": (26.47, 90.56),
            "Cachar": (24.83, 92.77), "Charaideo": (27.01, 94.88),
            "Chirang": (26.51, 90.62), "Darrang": (26.45, 92.03),
            "Dhemaji": (27.48, 94.58), "Dhubri": (26.02, 89.98),
            "Dibrugarh": (27.48, 94.90), "Dima Hasao": (25.19, 93.02),
            "Goalpara": (26.17, 90.62), "Golaghat": (26.52, 93.97),
            "Hailakandi": (24.68, 92.56), "Hojai": (26.00, 92.85),
            "Jorhat": (26.75, 94.22), "Kamrup": (26.32, 91.36),
            "Kamrup Metropolitan": (26.14, 91.73), "Karbi Anglong": (26.00, 93.50),
            "Karimganj": (24.86, 92.35), "Kokrajhar": (26.40, 90.27),
            "Lakhimpur": (27.23, 94.10), "Majuli": (26.96, 94.16),
            "Morigaon": (26.25, 92.33), "Nagaon": (26.35, 92.68),
            "Nalbari": (26.44, 91.44), "Sivasagar": (26.98, 94.63),
            "Sonitpur": (26.63, 92.83), "South Salmara-Mankachar": (25.72, 89.99),
            "Tinsukia": (27.50, 95.36), "Udalguri": (26.74, 92.09),
            "West Karbi Anglong": (25.98, 92.54),
        }

    def _get_dataset_hash(self):
        if self.df is None:
            return ""
        hash_str = f"{self.df.shape[0]}_{self.df.shape[1]}_{''.join(self.df.columns)}"
        return hashlib.sha256(hash_str.encode()).hexdigest()[:10]

    # ------------------------------------------------------------------
    # LOAD DATASET  —  now accepts a Gradio file object (temp path)
    # ------------------------------------------------------------------
    def load_dataset(self, file_obj):
        """Load dataset, populate ALL lookups, train models, return UI updates."""
        # Resolve file path from Gradio upload
        if file_obj is None:
            return ("❌ No file uploaded.",) + tuple(gr.update(choices=[], value=None) for _ in range(13))

        file_path = file_obj.name if hasattr(file_obj, 'name') else str(file_obj)
        file_path = file_path.strip()

        if not os.path.exists(file_path):
            return (f"❌ File not found: {file_path}",) + tuple(gr.update(choices=[], value=None) for _ in range(13))

        try:
            if file_path.lower().endswith('.csv'):
                self.df = pd.read_csv(file_path)
            else:
                self.df = pd.read_excel(file_path)
        except Exception as e:
            return (f"❌ Error reading file: {e}",) + tuple(gr.update(choices=[], value=None) for _ in range(13))

        REQUIRED = [
            "district", "crop_name", "crop_stage", "variety_name",
            "soil_type", "advisory_category", "advisory_priority"
        ]
        missing = [c for c in REQUIRED if c not in self.df.columns]
        if missing:
            return (f"❌ Schema Error: Missing columns → {missing}",) + tuple(gr.update(choices=[], value=None) for _ in range(13))

        # ---- Populate every lookup table (ALL before return) ----
        self.DISTRICTS = sorted(self.df["district"].unique().tolist())
        self.CROPS = sorted(self.df["crop_name"].unique().tolist())
        self.BLOCKS_BY_DISTRICT = {
            d: sorted(self.df[self.df["district"] == d]["block"].unique().tolist())
            for d in self.DISTRICTS
        }

        self.CROP_INFO = {}
        for crop in self.CROPS:
            sub = self.df[self.df["crop_name"] == crop]
            self.CROP_INFO[crop] = {
                "stages":    sorted(sub["crop_stage"].unique().tolist()),
                "varieties": sorted(sub["variety_name"].unique().tolist()),
                "families":  sub["crop_family"].mode()[0] if "crop_family" in sub and not sub.empty else "Other",
                "groups":    sub["crop_group"].mode()[0] if "crop_group" in sub and not sub.empty else "Other",
            }

        self.CATEGORY_TEXTS = {
            cat: self.df[self.df["advisory_category"] == cat]["advisory_text"].unique().tolist()
            for cat in self.df["advisory_category"].unique()
        }

        for attr, col in [
            ("SOIL_TYPES",         "soil_type"),
            ("WATER_SOURCES",      "water_source"),
            ("FIELD_CONDITIONS",   "field_condition"),
            ("DRAINAGE_LEVELS",    "plot_drainage"),
            ("PREVIOUS_CROPS",     "previous_crop"),
            ("PEST_HISTORIES",     "pest_disease_history"),
            ("FERTILIZERS",        "fertilizer_applied"),
            ("PESTICIDES",         "pesticide_applied"),
            ("IRRIGATION_SOURCES", "irrigation_source"),
            ("WATER_AVAIL_LEVELS", "water_availability_level"),
            ("WATERLOGGING_FREQS", "waterlogging_frequency"),
        ]:
            if col in self.df.columns:
                setattr(self, attr, sorted(self.df[col].unique().tolist()))
            else:
                setattr(self, attr, ["N/A"])

        for d in self.DISTRICTS:
            dsub = self.df[self.df["district"] == d]
            self.DISTRICT_INFO[d] = {
                "region":     dsub["region"].mode()[0]                    if "region" in dsub.columns and not dsub.empty else "Lower Assam",
                "zone":       dsub["agro_climatic_zone"].mode()[0]        if "agro_climatic_zone" in dsub.columns and not dsub.empty else "Central Zone",
                "flood_risk": dsub["district_base_flood_risk"].mode()[0]  if "district_base_flood_risk" in dsub.columns and not dsub.empty else "Moderate",
                "flood_zone": dsub["flood_hazard_zone"].mode()[0]         if "flood_hazard_zone" in dsub.columns and not dsub.empty else "Moderate",
            }

        num_cols = [
            "soil_ph", "soil_organic_carbon_pct", "soil_nitrogen_kg_ha",
            "soil_phosphorus_kg_ha", "soil_potassium_kg_ha",
            "soil_sulphur_mg_kg", "soil_zinc_mg_kg", "soil_boron_mg_kg",
        ]
        present_num = [c for c in num_cols if c in self.df.columns]
        if present_num:
            self.DISTRICT_SOIL_AVG = self.df.groupby("district")[present_num].mean().to_dict("index")

        self.data_loaded = True

        # Train models immediately (no separate click needed)
        train_msg = self._train_or_load_models()

        # Build Gradio dropdown updates
        def first(lst):
            return lst[0] if lst else None

        updates = [
            gr.update(choices=self.DISTRICTS,          value=first(self.DISTRICTS)),
            gr.update(choices=self.CROPS,              value=first(self.CROPS)),
            gr.update(choices=self.SOIL_TYPES,         value=first(self.SOIL_TYPES)),
            gr.update(choices=self.FIELD_CONDITIONS,   value=first(self.FIELD_CONDITIONS)),
            gr.update(choices=self.DRAINAGE_LEVELS,    value=first(self.DRAINAGE_LEVELS)),
            gr.update(choices=self.WATERLOGGING_FREQS, value=first(self.WATERLOGGING_FREQS)),
            gr.update(choices=self.WATER_SOURCES,      value=first(self.WATER_SOURCES)),
            gr.update(choices=self.IRRIGATION_SOURCES, value=first(self.IRRIGATION_SOURCES)),
            gr.update(choices=self.WATER_AVAIL_LEVELS, value=first(self.WATER_AVAIL_LEVELS)),
            gr.update(choices=self.PREVIOUS_CROPS,     value=first(self.PREVIOUS_CROPS)),
            gr.update(choices=self.PEST_HISTORIES,     value=first(self.PEST_HISTORIES)),
            gr.update(choices=self.FERTILIZERS,        value=first(self.FERTILIZERS)),
            gr.update(choices=self.PESTICIDES,         value=first(self.PESTICIDES)),
        ]

        return (f"✅ Dataset loaded. {train_msg}",) + tuple(updates)

    # ------------------------------------------------------------------
    # TRAIN / LOAD MODELS  —  with accuracy & overfitting checks
    # ------------------------------------------------------------------
    def _train_or_load_models(self):
        self.FEATURE_COLS = [
            c for c in self.df.columns
            if c not in [
                "farmer_id", "sowing_date", "advisory_category",
                "advisory_priority", "advisory_reason_code",
                "advisory_confidence", "advisory_text",
            ]
        ]
        self.CAT_COLS = self.df[self.FEATURE_COLS].select_dtypes(include=["object"]).columns.tolist()

        v_hash = self._get_dataset_hash()
        prefix = f"model_v10_{v_hash}"
        MODEL_FILES = [
            f"{prefix}_rf_cat.joblib", f"{prefix}_rf_pri.joblib",
            f"{prefix}_prep.joblib",   f"{prefix}_le_cat.joblib",
            f"{prefix}_le_pri.joblib", f"{prefix}_metrics.joblib",
        ]

        if all(os.path.exists(f) for f in MODEL_FILES):
            self.rf_cat, self.rf_pri, self.preprocessor, self.le_cat, self.le_pri, self.metrics = (
                joblib.load(f) for f in MODEL_FILES
            )
            self.is_ready = True
            return "Cached models loaded."

        print("   [!] Training ML models — this may take 1–2 minutes…")

        # Train / Test split (stratify to preserve class balance)
        X = self.df[self.FEATURE_COLS]
        y_cat = self.df["advisory_category"]
        y_pri = self.df["advisory_priority"]

        X_train, X_test, y_cat_train, y_cat_test, y_pri_train, y_pri_test = train_test_split(
            X, y_cat, y_pri, test_size=0.2, random_state=42, stratify=y_cat
        )

        # Preprocessing
        try:
            ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
        except TypeError:
            ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)

        self.preprocessor = ColumnTransformer(
            transformers=[("cat", ohe, self.CAT_COLS)], remainder="passthrough"
        )
        X_train_enc = self.preprocessor.fit_transform(X_train)
        X_test_enc = self.preprocessor.transform(X_test)

        self.le_cat = LabelEncoder().fit(y_cat_train)
        self.le_pri = LabelEncoder().fit(y_pri_train)

        y_cat_train_enc = self.le_cat.transform(y_cat_train)
        y_cat_test_enc = self.le_cat.transform(y_cat_test)
        y_pri_train_enc = self.le_pri.transform(y_pri_train)
        y_pri_test_enc = self.le_pri.transform(y_pri_test)

        # Random Forest with overfitting guard-rails
        self.rf_cat = RandomForestClassifier(
            n_estimators=150, max_depth=18, min_samples_split=5,
            min_samples_leaf=2, max_features='sqrt',
            n_jobs=-1, random_state=42
        ).fit(X_train_enc, y_cat_train_enc)

        self.rf_pri = RandomForestClassifier(
            n_estimators=150, max_depth=14, min_samples_split=5,
            min_samples_leaf=2, max_features='sqrt',
            n_jobs=-1, random_state=42
        ).fit(X_train_enc, y_pri_train_enc)

        # Evaluation helper
        def evaluate(model, X_tr, y_tr, X_te, y_te, name):
            tr_pred = model.predict(X_tr)
            te_pred = model.predict(X_te)
            tr_acc = accuracy_score(y_tr, tr_pred)
            te_acc = accuracy_score(y_te, te_pred)
            gap = tr_acc - te_acc

            if gap > 0.15:
                flag = "🔴 OVERFITTING"
            elif gap > 0.08:
                flag = "🟡 Mild overfit"
            else:
                flag = "🟢 Good generalization"

            return {
                "model": name,
                "train_accuracy": round(tr_acc, 4),
                "test_accuracy": round(te_acc, 4),
                "precision": round(precision_score(y_te, te_pred, average='weighted', zero_division=0), 4),
                "recall": round(recall_score(y_te, te_pred, average='weighted', zero_division=0), 4),
                "f1": round(f1_score(y_te, te_pred, average='weighted', zero_division=0), 4),
                "gap": round(gap, 4),
                "status": flag,
            }

        self.metrics = {
            "category": evaluate(self.rf_cat, X_train_enc, y_cat_train_enc, X_test_enc, y_cat_test_enc, "Category RF"),
            "priority": evaluate(self.rf_pri, X_train_enc, y_pri_train_enc, X_test_enc, y_pri_test_enc, "Priority RF"),
            "features": X_train_enc.shape[1],
            "train_n": len(X_train),
            "test_n": len(X_test),
        }

        for obj, path in zip(
            [self.rf_cat, self.rf_pri, self.preprocessor, self.le_cat, self.le_pri, self.metrics],
            MODEL_FILES,
        ):
            joblib.dump(obj, path)

        self.is_ready = True
        m = self.metrics
        return (f"Models trained. "
                f"Category Acc={m['category']['test_accuracy']:.1%} ({m['category']['status']}) | "
                f"Priority Acc={m['priority']['test_accuracy']:.1%} ({m['priority']['status']})")

engine = AssamAIEngine()

# --- 3. WEATHER & PEST ENGINES (unchanged logic, robust wrappers) ---
WMO_CODES = {
    0: "Clear", 1: "Mainly clear", 2: "Partly cloudy", 3: "Overcast",
    45: "Fog", 51: "Drizzle", 61: "Slight rain", 63: "Moderate rain",
    65: "Heavy rain", 95: "Thunderstorm", 99: "Hailstorm",
}

def get_weather_forecast(lat, lon):
    url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": lat, "longitude": lon,
        "current": ["temperature_2m", "relative_humidity_2m", "wind_speed_10m", "weather_code"],
        "daily":   ["temperature_2m_max", "temperature_2m_min", "precipitation_sum",
                    "wind_speed_10m_max", "weather_code"],
        "timezone": "auto", "forecast_days": 7,
    }
    try:
        return retry_session.get(url, params=params, timeout=20).json()
    except Exception:
        return None

def analyze_7day_weather(wa):
    if not wa:
        return None
    daily = wa.get("daily", {})
    days = []
    sudden_shift_alert = ""

    n_days = min(7, len(daily.get("time", [])))
    for i in range(n_days):
        r = daily["precipitation_sum"][i]
        t = daily["temperature_2m_max"][i]
        w = daily["wind_speed_10m_max"][i]
        c = daily["weather_code"][i]

        rain_3day = sum(daily["precipitation_sum"][i: i + 3]) if i + 3 <= n_days else sum(daily["precipitation_sum"][i:])

        if c in [95, 96]:    cond = "Thunderstorm"
        elif c == 99:        cond = "Hailstorm"
        elif r > 50:         cond = "Heavy Rain"
        elif r > 5:          cond = "Rainy"
        elif r == 0 and t > 35: cond = "Heatwave"
        elif r == 0 and t > 33: cond = "Dry Spell"
        else:                cond = "Clear"

        days.append({
            "date": daily["time"][i], "rain_mm": r, "rain_3day": rain_3day,
            "max_temp": t, "min_temp": daily["temperature_2m_min"][i],
            "wind_kmh": w, "humidity": wa.get("current", {}).get("relative_humidity_2m", 70),
            "condition": cond, "wmo_code": c,
        })

    if len(days) >= 2:
        if days[1]["rain_mm"] - days[0]["rain_mm"] > 30:
            sudden_shift_alert = (
                f"⚠️ SUDDEN WEATHER SHIFT: Heavy rain ({days[1]['rain_mm']}mm) "
                f"expected tomorrow following today's ({days[0]['rain_mm']}mm). Act now!"
            )
        elif days[0]["max_temp"] - days[1]["max_temp"] > 8:
            sudden_shift_alert = (
                f"⚠️ TEMPERATURE DROP: Sudden "
                f"{days[0]['max_temp'] - days[1]['max_temp']}°C drop expected tomorrow. "
                f"Protect nurseries."
            )

    return {
        "days": days,
        "total_rain":    sum(daily.get("precipitation_sum", [])),
        "max_rain_3day": sum(daily.get("precipitation_sum", [])[:3]),
        "max_temp_week": max(daily.get("temperature_2m_max", [30])),
        "min_temp_week": min(daily.get("temperature_2m_min", [20])),
        "max_wind_week": max(daily.get("wind_speed_10m_max", [0])),
        "avg_humidity":  wa.get("current", {}).get("relative_humidity_2m", 70),
        "thunder_days":  [d for d in days if "Thunder" in d["condition"]],
        "hail_days":     [d for d in days if "Hail" in d["condition"]],
        "sudden_shift":  sudden_shift_alert,
    }

def calculate_risk_summary(wa_subset, dist_info, crop, stage, soil, f_cond, drain, w_av):
    fs = ds = hs = ss = 0
    mult = {"Low": 1, "Moderate": 1.5, "High": 2, "Very High": 3}.get(
        dist_info.get("flood_risk", "Moderate"), 1
    )

    rain_val = wa_subset.get("rain_3day",
               wa_subset.get("max_rain_3day",
               wa_subset.get("rain_mm", 0)))
    if rain_val > 150: fs = 90
    elif rain_val > 100: fs = 70
    elif rain_val > 50:  fs = 50
    elif rain_val > 20:  fs = 25
    fs = min(100, int(fs * mult))
    if "Lowland" in str(f_cond): fs = min(100, fs + 15)
    if drain in ["Poor", "Very Poor"]: fs = min(100, fs + 10)

    rain_total = wa_subset.get("total_rain", wa_subset.get("rain_mm", 0))
    temp_val   = wa_subset.get("max_temp_week", wa_subset.get("max_temp", 30))
    if rain_total < 10 and temp_val > 33: ds = 60
    elif rain_total < 25: ds = 30
    if "Scarce" in str(w_av): ds = min(100, ds + 15)
    if "Sandy" in str(soil): ds = min(100, ds + 10)

    if temp_val >= 38:   hs = 80
    elif temp_val >= 35: hs = 55
    if stage in ["Flowering", "Grain filling"]: hs = min(100, hs + 10)
    if crop in ["Wheat", "Mustard", "Potato", "Tea"]: hs = min(100, hs + 5)

    hail    = len(wa_subset.get("hail_days", [])) > 0 if "hail_days" in wa_subset else ("Hail" in wa_subset.get("condition", ""))
    thunder = len(wa_subset.get("thunder_days", [])) >= 1 if "thunder_days" in wa_subset else ("Thunder" in wa_subset.get("condition", ""))
    wind_val = wa_subset.get("max_wind_week", wa_subset.get("wind_kmh", 10))

    if hail:        ss = 95
    elif thunder:   ss = 70
    elif wind_val >= 40: ss = 60
    if crop in ["Banana", "Tea", "Maize", "Sugarcane"]: ss = min(100, ss + 5)

    def s_lvl(s):
        return ("CRITICAL", "🔴") if s >= 75 else \
               ("HIGH",     "🟠") if s >= 50 else \
               ("MODERATE", "🟡") if s >= 25 else \
               ("LOW",      "🟢")

    return {
        "flood":     {"score": fs, "level": s_lvl(fs)[0], "icon": s_lvl(fs)[1]},
        "drought":   {"score": ds, "level": s_lvl(ds)[0], "icon": s_lvl(ds)[1]},
        "heat":      {"score": hs, "level": s_lvl(hs)[0], "icon": s_lvl(hs)[1]},
        "storm":     {"score": ss, "level": s_lvl(ss)[0], "icon": s_lvl(ss)[1]},
        "max_score": max(fs, ds, hs, ss),
    }

def predict_pest_disease_risk(wa, crop, stage, p_hist, soil, f_cond, drain):
    days = wa["days"]
    r_score = 0
    factors, issues = [], []

    humid_days = len([d for d in days if d["humidity"] > 80 and 20 <= d["max_temp"] <= 30])
    if humid_days >= 3:
        r_score += 25
        factors.append(f"{humid_days} warm-humid days (fungal risk)")
        if "Rice" in crop:       issues.extend(["Rice Blast", "Sheath Blight"])
        elif crop in ["Potato", "Tomato"]: issues.append("Late Blight (Phytophthora)")
        elif crop == "Tea":      issues.append("Blister Blight")
        elif crop == "Mustard":  issues.append("Alternaria Blight")
        elif crop in ["Ginger", "Turmeric"]: issues.append("Rhizome Rot / Leaf Spot")
        elif crop in ["Green Gram", "Black Gram", "Lentil", "Chickpea"]: issues.append("Powdery Mildew / Wilt")
        elif crop == "Sugarcane": issues.append("Red Rot")
        elif crop in ["Coconut", "Arecanut"]: issues.append("Bud Rot")

    dry_days = len([d for d in days if d["rain_mm"] < 1 and d["max_temp"] > 33])
    if dry_days >= 3:
        r_score += 15
        factors.append(f"{dry_days} hot-dry days (sucking pests)")
        if crop == "Tea":        issues.append("Red Spider Mite")
        elif crop in ["Tomato", "Cotton", "Okra", "Brinjal"]: issues.append("Whitefly / Aphids")
        elif "Rice" in crop and stage in ["Vegetative"]: issues.append("Yellow Stem Borer")
        elif crop == "Mustard":  issues.append("Mustard Aphid")
        elif crop in ["Lentil", "Chickpea"]: issues.append("Pod Borer")

    heavy_rain = len([d for d in days if d["rain_mm"] >= 40])
    if heavy_rain >= 1 and ("Poor" in str(drain) or "Lowland" in str(f_cond)):
        r_score += 20
        factors.append("Heavy rain + poor drainage (bacterial/rot risk)")
        if "Rice" in crop:   issues.append("Bacterial Leaf Blight (BLB)")
        elif crop in ["Tomato", "Banana"]: issues.append("Bacterial Wilt")
        elif crop == "Jute": issues.append("Stem Rot")
        elif crop in ["Ginger", "Turmeric"]: issues.append("Soft Rot")

    if p_hist and p_hist != "No significant pest/disease":
        r_score += 15
        factors.append(f"History: {p_hist}")

    r_score = min(100, r_score)
    lvl = "HIGH" if r_score >= 60 else ("MODERATE" if r_score >= 35 else "LOW")
    return {
        "score": r_score, "level": lvl,
        "icon":  "🔴" if lvl == "HIGH" else "🟡" if lvl == "MODERATE" else "🟢",
        "factors": factors, "issues": list(set(issues)),
    }

# --- 4. ML FEATURE BUILDER ---
def build_feature_row(dist, block, vill, crop, stage, var, soil,
                      w_src, irrig, w_av, f_cond, drain, w_log,
                      p_crop, p_hist, fert, pest, age, day):
    d_info = engine.DISTRICT_INFO.get(dist, {"region": "Lower Assam", "zone": "Central Zone", "flood_risk": "Moderate", "flood_zone": "Moderate"})
    c_info = engine.CROP_INFO.get(crop, {"families": "Cereals", "groups": "Kharif"})
    s_avg  = engine.DISTRICT_SOIL_AVG.get(dist, {})

    texture_map = {
        "Alluvial (Brahmaputra)": "Loamy", "Loamy": "Loamy", "Sandy": "Sandy",
        "Clayey": "Clay", "Red & Yellow": "Sandy loam",
        "Peaty/Marshy": "Clay loam", "Laterite": "Sandy clay loam",
    }
    texture = texture_map.get(str(soil), "Loamy")

    ph       = s_avg.get("soil_ph", 5.5)
    ph_class = "Strongly acidic" if ph < 5.5 else ("Moderately acidic" if ph < 6.5 else "Neutral")
    oc       = s_avg.get("soil_organic_carbon_pct", 0.7)
    oc_class = "Low" if oc < 0.5 else ("Medium" if oc < 0.75 else "High")
    n_val    = s_avg.get("soil_nitrogen_kg_ha", 220)
    n_class  = "Low" if n_val < 200 else ("Medium" if n_val < 300 else "High")

    stage_rank = {"Nursery": 1, "Sowing": 2, "Transplanting": 3, "Vegetative": 4,
                  "Flowering": 5, "Grain filling": 6, "Harvest": 7}.get(stage, 4)
    season = {12: "Winter", 1: "Winter", 2: "Winter", 3: "Spring", 4: "Spring",
              5: "Spring", 6: "Summer", 7: "Summer", 8: "Summer",
              9: "Autumn", 10: "Autumn", 11: "Autumn"}.get(datetime.now().month, "Summer")

    smi = min(100, max(0,
        oc * 30 + (day["rain_3day"] / 5) + (day["humidity"] / 2)
        - (day["max_temp"] - 25) * 2
    ))

    return {
        "district": dist, "block": block, "village": vill,
        "region": d_info["region"], "agro_climatic_zone": d_info["zone"],
        "district_base_flood_risk": d_info["flood_risk"], "flood_hazard_zone": d_info["flood_zone"],
        "crop_name": crop, "crop_family": c_info["families"], "crop_group": c_info["groups"],
        "crop_stage": stage, "crop_stage_rank": stage_rank, "crop_age_days": age,
        "forecast_month": datetime.now().month, "forecast_season": season,
        "variety_name": var, "soil_type": soil, "soil_texture_class": texture,
        "soil_ph": ph, "soil_ph_class": ph_class,
        "soil_organic_carbon_pct": oc, "soil_oc_class": oc_class,
        "soil_nitrogen_kg_ha": n_val, "soil_n_class": n_class,
        "soil_phosphorus_kg_ha": s_avg.get("soil_phosphorus_kg_ha", 18), "soil_p_class": "Medium",
        "soil_potassium_kg_ha":  s_avg.get("soil_potassium_kg_ha", 160), "soil_k_class": "Medium",
        "soil_sulphur_mg_kg":    s_avg.get("soil_sulphur_mg_kg", 12),    "soil_s_class": "Medium",
        "soil_zinc_mg_kg":       s_avg.get("soil_zinc_mg_kg", 0.8),      "soil_zn_class": "Adequate",
        "soil_boron_mg_kg":      s_avg.get("soil_boron_mg_kg", 0.5),     "soil_b_class": "Adequate",
        "water_source": w_src, "irrigation_source": irrig,
        "water_availability_level": w_av, "field_condition": f_cond,
        "plot_drainage": drain, "waterlogging_frequency": w_log,
        "previous_crop": p_crop, "pest_disease_history": p_hist,
        "fertilizer_applied": fert, "pesticide_applied": pest,
        "weather_forecast_rain_3day_mm": day["rain_3day"],
        "weather_forecast_max_temp_c":   day["max_temp"],
        "weather_forecast_min_temp_c":   day["min_temp"],
        "weather_forecast_humidity_pct": day["humidity"],
        "weather_forecast_wind_kmh":     day["wind_kmh"],
        "weather_condition": day["condition"], "soil_moisture_index": smi,
    }

# --- 5. ADVISORY GENERATOR ---
def get_deterministic_text(category, fingerprint, texts):
    if not texts:
        return "Monitor your crop regularly."
    idx = int(hashlib.md5(fingerprint.encode()).hexdigest(), 16) % len(texts)
    return texts[idx]

def generate_full_natural_advisory(row, day, dist, dist_info):
    # Safety: ensure every expected feature exists
    r_df = pd.DataFrame([row])
    for col in engine.FEATURE_COLS:
        if col not in r_df.columns:
            if col in engine.CAT_COLS:
                r_df[col] = engine.df[col].mode()[0] if not engine.df[col].empty else "Unknown"
            else:
                r_df[col] = engine.df[col].median() if not engine.df[col].empty else 0

    r_enc = engine.preprocessor.transform(r_df[engine.FEATURE_COLS])

    pred_cat_idx = engine.rf_cat.predict(r_enc)[0]
    cat  = engine.le_cat.inverse_transform([pred_cat_idx])[0]
    pri  = engine.le_pri.inverse_transform([engine.rf_pri.predict(r_enc)[0]])[0]
    conf = round(engine.rf_cat.predict_proba(r_enc)[0][pred_cat_idx], 3)

    f_print = f"{row['district']}{row['crop_name']}{row['crop_stage']}{row['soil_type']}{day['condition']}"
    base = get_deterministic_text(cat, f_print, engine.CATEGORY_TEXTS.get(cat, ["Monitor crop."]))

    why = alt = res = lt = ""

    advisory_map = {
        "Flood Emergency": (
            f"SEVERE ALERT: {day['rain_mm']}mm rain expected. "
            "1) Harvest mature crops today. 2) Move livestock/pumps. "
            "3) Open all drainage outlets. 4) Do not enter flooded fields.",
            f"{day['rain_mm']}mm rain in 24h + {dist_info.get('flood_risk','High')} risk + {row['field_condition']} = critical waterlogging.",
            "Cover mature crops with waterproof tarpaulin and raise on bamboo platforms.",
            "Sandbags, shovels, waterproof bags, emergency contacts.",
            "Construct raised platforms. Use submergence-tolerant varieties next season.",
        ),
        "Flood Watch": (
            "Flood watch active. Clear drainage channels, secure bunds, and avoid fertilizer application for 48 hours.",
            f"{day['rain_mm']}mm rain + {row['plot_drainage']} drainage = potential inundation.",
            "Create temporary furrows every 10 meters.",
            "Sandbags, spades for channels.",
            "Improve field drainage with underground pipes.",
        ),
        "Drainage Action": (
            "Heavy soil saturation detected. Lower field water level to ensure no stagnation beyond 12 hours.",
            f"{row['soil_type']} soil + {day['rain_mm']}mm rain = root oxygen starvation.",
            "Apply gypsum @ 100kg/acre to improve percolation if drainage fails.",
            "Spades, gypsum, sand.",
            "Adopt raised-bed cultivation.",
        ),
        "Drought Watch": (
            "Dry conditions ahead. Budget water carefully and apply organic mulch.",
            f"{day['max_temp']}°C heat + {row['water_availability_level']} water = moisture stress.",
            "Irrigate only at critical flowering/grain-filling stages.",
            "Paddy straw mulch, black polythene.",
            "Dig farm pond for rainwater harvesting.",
        ),
        "Irrigation Advisory": (
            "Irrigation recommended. Irrigate early morning (5–7 AM) or evening to minimize evaporative loss.",
            f"Soil moisture depleted + {row['crop_stage']} stage requires water.",
            "If canal water unavailable, irrigate at night using stored water.",
            "Water pump, hose pipes.",
            "Switch to drip/sprinkler for 40% water savings.",
        ),
        "Heat Stress Protection": (
            "Heat wave alert. Apply light irrigation during peak hours. Avoid chemical spraying.",
            f"{day['max_temp']}°C exceeds optimal range for {row['crop_name']}.",
            "Spray kaolin clay solution to reflect heat.",
            "Shade nets (50%), kaolin clay.",
            "Plant windbreaks. Apply mulch.",
        ),
        "Cold Protection": (
            "Cold snap expected. Increase standing water depth to act as a thermal buffer.",
            f"{day['min_temp']}°C below threshold = frost damage risk.",
            "Create light smoke at dawn for thermal protection.",
            "Dry straw, polythene sheets.",
            "Use cold-tolerant varieties.",
        ),
        "Disease Prevention": (
            "Warm-humid conditions favor fungal diseases. Apply preventive spray of Trichoderma/Bordeaux mixture.",
            f"{day['humidity']}% humidity + {day['max_temp']}°C creates fungal breeding ground.",
            "Increase plant spacing for airflow. Spray neem oil.",
            "Trichoderma, Bordeaux mixture.",
            "Practice crop rotation.",
        ),
        "Harvest Now": (
            "OPTIMAL HARVEST WINDOW. Harvest immediately before weather deteriorates.",
            f"Stable {day['condition']} weather is optimal before next rain.",
            "Hire contract harvesters to speed up process.",
            "Sickles, gunny bags, tarpaulin.",
            "Build proper ventilated storage.",
        ),
        "Spraying Window": (
            "Ideal spraying conditions today. Apply necessary foliar sprays early morning.",
            f"Wind {day['wind_kmh']}km/h (safe) + no rain washout expected.",
            "If rain suddenly appears, use sticker-spreader.",
            "Knapsack sprayer, protective gear.",
            "Install pheromone traps.",
        ),
        "Thunderstorm Safety": (
            "THUNDERSTORM EXPECTED. Lightning risk is HIGH. Do not work in open fields.",
            "Thunderstorm forecast + lightning risk in open fields.",
            "Document any crop damage for insurance.",
            "Rope for securing structures.",
            "Install lightning conductors on sheds.",
        ),
        "Wind Protection": (
            f"Strong winds forecast. Install bamboo supports and secure polyhouses.",
            f"{day['wind_kmh']}km/h wind causes lodging risk for {row['crop_name']}.",
            "Tie plants in bundles with jute rope.",
            "Bamboo poles, jute rope.",
            "Plant Gliricidia windbreaks.",
        ),
        "Hailstorm Protection": (
            "HAILSTORM risk! Cover crops with agro-net if possible.",
            "Hailstorm physical damage risk.",
            "Apply Bordeaux paste on cuts post-hail to prevent infection.",
            "Agro-net (40% shade).",
            "Get PMFBY crop insurance.",
        ),
        "Fertility Management": (
            "Soil nutrient status suggests fertilizer application. Apply balanced NPK.",
            f"{row['crop_stage']} is peak nutrient demand phase.",
            "Apply vermicompost if chemical NPK unavailable.",
            "Urea/DAP, soil test kit.",
            "Adopt Integrated Nutrient Management.",
        ),
        "Soil Moisture Conservation": (
            "Conserve every drop. Use mulching and contour bunding.",
            f"Scarcity ({day['rain_mm']}mm) + {day['max_temp']}°C evaporation.",
            "Use pitcher irrigation.",
            "Mulch material.",
            "Install rainwater harvesting.",
        ),
    }

    for key in ["Sowing Window", "Transplanting Window"]:
        advisory_map[key] = (
            "EXCELLENT sowing/transplanting conditions. Prepare seedbed with fine tilth.",
            f"{day['condition']} + optimal soil moisture ensures maximum germination.",
            "Treat seeds with Trichoderma if untreated.",
            "Quality seeds, basal fertilizer.",
            "Maintain a seed bank.",
        )

    if cat in advisory_map:
        base, why, alt, res, lt = advisory_map[cat]
    else:
        why = f"All parameters optimal for {row['crop_name']} at {row['crop_stage']}."
        alt = "Prioritize pending farm maintenance."
        res = "Standard farm tools."
        lt  = "Maintain daily farm diary."

    return {"cat": cat, "pri": pri, "conf": conf,
            "main": base, "why": why, "alt": alt, "res": res, "lt": lt}

# --- 6. MASTER PROCESS (single, clean definition) ---
def _risk_bg(level):
    return {"CRITICAL": "#c0392b", "HIGH": "#e67e22",
            "MODERATE": "#f1c40f", "LOW": "#27ae60"}.get(level, "#27ae60")

def master_process(dist, block, vill, crop, stage, var, sow_str,
                   soil, w_src, irrig, w_av, f_cond, drain, w_log,
                   p_crop, p_hist, fert, pest):
    if not engine.data_loaded:
        return None, None, "<h3 style='color:red;'>❌ Please upload a dataset first.</h3>"

    if not engine.is_ready:
        try:
            engine._train_or_load_models()
        except Exception as e:
            return None, None, f"<h3 style='color:red;'>❌ Model training failed: {e}</h3>"

    try:
        sow_date = datetime.strptime(sow_str, "%Y-%m-%d").date()
    except Exception:
        sow_date = (datetime.now() - timedelta(days=60)).date()
    age = (datetime.now().date() - sow_date).days

    if stage == "Harvest" and age < 60:
        return None, None, (
            f"<h3 style='color:red;'>⚠️ Conflict: Sowing date ({age} days ago) "
            "is too recent for Harvest stage.</h3>"
        )

    coords  = engine.DISTRICT_COORDS.get(dist, (26.14, 91.73))
    wa_raw  = get_weather_forecast(coords[0], coords[1])
    if not wa_raw:
        return None, None, "<h3 style='color:red;'>❌ API Error: Could not fetch weather data. Check internet.</h3>"

    wa        = analyze_7day_weather(wa_raw)
    dist_info = engine.DISTRICT_INFO.get(dist, {"region": "Lower Assam", "flood_risk": "Moderate"})

    rs = calculate_risk_summary(wa, dist_info, crop, stage, soil, f_cond, drain, w_av)
    pr = predict_pest_disease_risk(wa, crop, stage, p_hist, soil, f_cond, drain)

    advisories = []
    for day in wa["days"]:
        row = build_feature_row(
            dist, block, vill, crop, stage, var, soil,
            w_src, irrig, w_av, f_cond, drain, w_log,
            p_crop, p_hist, fert, pest, age, day
        )
        adv = generate_full_natural_advisory(row, day, dist, dist_info)
        adv["day"] = day
        advisories.append(adv)

    # ── ML Metrics Panel ───────────────────────────────────────────
    m = engine.metrics
    metrics_html = ""
    if m:
        metrics_html = f"""
        <div style="background:#f8f9fa;padding:12px;border-radius:6px;margin-bottom:15px;border:1px solid #dee2e6;">
          <h4 style="margin-top:0;color:#1a5276;">🤖 ML Model Health (Random Forest)</h4>
          <table style="width:100%;font-size:13px;border-collapse:collapse;">
            <tr style="background:#e9ecef;"><th style="text-align:left;padding:6px;">Metric</th><th style="text-align:left;padding:6px;">Category Model</th><th style="text-align:left;padding:6px;">Priority Model</th></tr>
            <tr><td style="padding:6px;border-bottom:1px solid #dee2e6;"><b>Test Accuracy</b></td><td style="padding:6px;border-bottom:1px solid #dee2e6;">{m['category']['test_accuracy']:.1%}</td><td style="padding:6px;border-bottom:1px solid #dee2e6;">{m['priority']['test_accuracy']:.1%}</td></tr>
            <tr><td style="padding:6px;border-bottom:1px solid #dee2e6;"><b>F1 Score</b></td><td style="padding:6px;border-bottom:1px solid #dee2e6;">{m['category']['f1']:.3f}</td><td style="padding:6px;border-bottom:1px solid #dee2e6;">{m['priority']['f1']:.3f}</td></tr>
            <tr><td style="padding:6px;border-bottom:1px solid #dee2e6;"><b>Precision</b></td><td style="padding:6px;border-bottom:1px solid #dee2e6;">{m['category']['precision']:.3f}</td><td style="padding:6px;border-bottom:1px solid #dee2e6;">{m['priority']['precision']:.3f}</td></tr>
            <tr><td style="padding:6px;border-bottom:1px solid #dee2e6;"><b>Recall</b></td><td style="padding:6px;border-bottom:1px solid #dee2e6;">{m['category']['recall']:.3f}</td><td style="padding:6px;border-bottom:1px solid #dee2e6;">{m['priority']['recall']:.3f}</td></tr>
            <tr><td style="padding:6px;border-bottom:1px solid #dee2e6;"><b>Train→Test Gap</b></td><td style="padding:6px;border-bottom:1px solid #dee2e6;">{m['category']['gap']:.2%} {m['category']['status']}</td><td style="padding:6px;border-bottom:1px solid #dee2e6;">{m['priority']['gap']:.2%} {m['priority']['status']}</td></tr>
          </table>
          <p style="margin:8px 0 0 0;font-size:11px;color:#6c757d;">
            Models: <b>RandomForestClassifier</b> (150 trees, depth 18/14, sqrt features).
            Samples: {m['train_n']} train / {m['test_n']} test. Features: {m['features']}.
          </p>
        </div>
        """

    # ── HTML report ─────────────────────────────────────────────────
    html = f"""
    <div style="font-family:sans-serif;border:1px solid #ddd;border-radius:10px;overflow:hidden;">
      <div style="background:linear-gradient(135deg,#1a5276,#2ecc71);color:white;padding:20px;">
        <h2 style="margin:0;">🌾 Assam Farmer AI v10.0 Dashboard</h2>
        <p style="margin:5px 0 0 0;">Report for <b>{dist}</b> | {crop} ({stage}) | Age: {age} days</p>
      </div>
    """

    if wa.get("sudden_shift"):
        html += f"<div style='background:#e74c3c;color:white;padding:12px;text-align:center;font-size:16px;'><b>{wa['sudden_shift']}</b></div>"

    html += f"""
      <div style="padding:20px;">
        {metrics_html}
        <h3 style="color:#c0392b;">🧠 Risk Summary</h3>
        <div style="display:flex;gap:10px;text-align:center;color:white;">
          <div style="flex:1;background:{_risk_bg(rs['flood']['level'])};padding:10px;border-radius:5px;">🌊 Flood<br><b>{rs['flood']['level']} ({rs['flood']['score']}/100)</b></div>
          <div style="flex:1;background:{_risk_bg(rs['drought']['level'])};padding:10px;border-radius:5px;">🏜️ Drought<br><b>{rs['drought']['level']} ({rs['drought']['score']}/100)</b></div>
          <div style="flex:1;background:{_risk_bg(rs['heat']['level'])};padding:10px;border-radius:5px;">🔥 Heat<br><b>{rs['heat']['level']} ({rs['heat']['score']}/100)</b></div>
          <div style="flex:1;background:{_risk_bg(rs['storm']['level'])};padding:10px;border-radius:5px;">⛈️ Storm<br><b>{rs['storm']['level']} ({rs['storm']['score']}/100)</b></div>
        </div>
        <hr>
        <h3 style="color:#c0392b;">🐛 Pest & Disease Matrix</h3>
        <div style="background:#fff5f5;padding:15px;border-left:5px solid {'#c0392b' if pr['level']=='HIGH' else '#27ae60'};">
          <p><b>Risk Level:</b> {pr['level']} ({pr['score']}/100)</p>
          <p><b>Watch for:</b> {', '.join(pr['issues']) if pr['issues'] else 'No major issues predicted.'}</p>
          <p><b>Factors:</b> {', '.join(pr['factors']) if pr['factors'] else 'Stable conditions.'}</p>
        </div>
        <hr>
        <h3>🚨 Priority Action Item</h3>
        <div style="background:#f4f6f7;padding:15px;border-left:5px solid {'#c0392b' if advisories[0]['pri'] in ['Critical','High'] else '#27ae60'};">
          <b style="color:#c0392b;">{advisories[0]['cat'].upper()} ({advisories[0]['pri']}):</b> {advisories[0]['main']}
        </div>
        <hr>
        <h3>📅 7-Day Action Calendar</h3>
    """

    for adv in advisories:
        d = adv["day"]
        html += f"""
        <div style="margin-bottom:10px;padding:10px;background:#f9f9f9;border-left:4px solid {'#c0392b' if adv['pri'] in ['Critical','High'] else '#3498db'};">
          <b>{d['date']} | {d['condition']} ({d['rain_mm']}mm, {d['max_temp']}°C)</b>
          <span style="float:right;font-size:11px;background:#e0e0e0;padding:2px 6px;border-radius:4px;">{adv['cat']}</span><br>
          {adv['main']}<br>
          <div style="margin-top:5px;font-size:12px;color:#555;background:#fff;padding:5px;border-radius:4px;">
            <b>Why:</b> {adv['why']}<br>
            <b>Alternative:</b> {adv['alt']}<br>
            <b>Resources:</b> {adv['res']}<br>
            <b>Long-Term:</b> {adv['lt']}
          </div>
        </div>
        """

    html += f"""
        <hr>
        <h3>🌱 Crop Specifics</h3>
        <p>Ensure management practices are aligned with <b>{crop}</b> requirements during the <b>{stage}</b> phase.</p>
        <h3>💧 Soil + Water Advice</h3>
        <p>Your <b>{soil}</b> soil has <b>{drain}</b> drainage. Keep water source ({w_src}) ready — {wa['total_rain']:.1f}mm forecast over 7 days.</p>
        <h3>🌾 Operations</h3>
        <p>Fertilizer: {fert}. Pesticide: {pest}. Consult 7-day calendar before next application.</p>
        <h3>📍 Localized Context</h3>
        <p>{dist} | {f_cond} terrain. District base flood risk: {dist_info.get('flood_risk','Moderate')}.</p>
        <h3>📉 ML Confidence Score</h3>
        <p><b style="color:{'red' if advisories[0]['conf'] < 0.65 else 'green'};">{advisories[0]['conf']:.1%}</b>
        {'⚠️ Low confidence — verify with KVK.' if advisories[0]['conf'] < 0.65 else '✅ High confidence via ML.'}</p>
      </div>
    </div>
    """

    # ── Charts ──────────────────────────────────────────────────────
    plt.close("all")
    days_labels = [d["date"][5:] for d in wa["days"]]

    fig1, ax1 = plt.subplots(figsize=(8, 3))
    ax1.bar(days_labels, [d["rain_mm"] for d in wa["days"]], color="skyblue", label="Rain (mm)")
    ax2 = ax1.twinx()
    ax2.plot(days_labels, [d["max_temp"] for d in wa["days"]], color="red", marker="o", label="Max Temp °C")
    ax1.set_ylabel("Rainfall (mm)")
    ax2.set_ylabel("Temperature (°C)")
    fig1.legend(loc="upper left")
    ax1.set_title("7-Day Weather Trend")
    plt.tight_layout()

    fig2, ax_heat = plt.subplots(figsize=(8, 1.5))
    daily_risks = [
        calculate_risk_summary(d, dist_info, crop, stage, soil, f_cond, drain, w_av)["max_score"]
        for d in wa["days"]
    ]
    cmap   = LinearSegmentedColormap.from_list("rg", ["#27ae60", "#f1c40f", "#e67e22", "#c0392b"], N=256)
    colors = [cmap(r / 100) for r in daily_risks]
    ax_heat.barh(["Risk"], [1] * len(days_labels), color=colors, edgecolor="white", height=1)
    for i, (date, risk) in enumerate(zip(days_labels, daily_risks)):
        ax_heat.text(i + 0.5, 0, f"{date}\n{risk}", ha="center", va="center",
                     color="white", fontweight="bold", fontsize=10)
    ax_heat.set_xlim(0, len(days_labels))
    ax_heat.axis("off")
    ax_heat.set_title("Daily Multi-Factor Hazard Heatmap (0–100)", pad=10)
    plt.tight_layout()

    return fig1, fig2, html


# --- 7. GRADIO UI ---
print("[4/4] Building Gradio interface…")

with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 🌾 Assam Farmer AI (v10.0 — Bulletproof Edition)")
    gr.Markdown(
        "### Step 1 — Upload your dataset\n"
        "Click **Upload Dataset** below, choose your `.csv` or `.xlsx` file, then click **Load Dataset**. "
        "All dropdowns will populate automatically and models will train instantly."
    )

    with gr.Row():
        file_upload = gr.File(
            label="Upload Dataset (CSV or XLSX)",
            file_types=[".csv", ".xlsx"],
        )
        load_btn = gr.Button("📂 Load Dataset", variant="secondary")
        init_status = gr.Textbox(
            label="System Status", interactive=False,
            placeholder="Upload a file and click Load Dataset…",
        )

    with gr.Row():
        # ── Left column: inputs ─────────────────────────────────────
        with gr.Column(scale=1):
            with gr.Accordion("Step 2: Location & Crop", open=True):
                dist_in  = gr.Dropdown(choices=[], label="District",  interactive=True)
                block_in = gr.Dropdown(choices=[], label="Block",     interactive=True)
                vill_in  = gr.Textbox("Main Village", label="Village")
                crop_in  = gr.Dropdown(choices=[], label="Crop",      interactive=True)
                stage_in = gr.Dropdown(choices=[], label="Stage",     interactive=True)
                var_in   = gr.Dropdown(choices=[], label="Variety",   interactive=True)
                sow_in   = gr.Textbox(
                    (datetime.now() - timedelta(days=60)).strftime("%Y-%m-%d"),
                    label="Sowing Date (YYYY-MM-DD)"
                )
            with gr.Accordion("Step 3: Soil & Field", open=False):
                soil_in   = gr.Dropdown(choices=[], label="Soil Type",       interactive=True)
                f_cond_in = gr.Dropdown(choices=[], label="Field Condition", interactive=True)
                drain_in  = gr.Dropdown(choices=[], label="Drainage",        interactive=True)
                w_log_in  = gr.Dropdown(choices=[], label="Waterlogging",    interactive=True)
            with gr.Accordion("Step 4: Water & History", open=False):
                w_src_in  = gr.Dropdown(choices=[], label="Water Source",   interactive=True)
                irrig_in  = gr.Dropdown(choices=[], label="Irrigation",     interactive=True)
                w_av_in   = gr.Dropdown(choices=[], label="Availability",   interactive=True)
                p_crop_in = gr.Dropdown(choices=[], label="Prev Crop",      interactive=True)
                p_hist_in = gr.Dropdown(choices=[], label="Pest History",   interactive=True)
                fert_in   = gr.Dropdown(choices=[], label="Fertilizer",     interactive=True)
                pest_in   = gr.Dropdown(choices=[], label="Pesticide",      interactive=True)
            btn = gr.Button("🧠 GENERATE COMPLETE ADVISORY", variant="primary")

        # ── Right column: outputs ───────────────────────────────────
        with gr.Column(scale=2):
            out_plot1 = gr.Plot(label="📊 Weather Visualization")
            out_plot2 = gr.Plot(label="🔥 Daily Risk Heatmap")
            out_html  = gr.HTML()

    # Dropdown list for single-shot update
    ALL_DROPDOWNS = [
        dist_in, crop_in, soil_in, f_cond_in, drain_in, w_log_in,
        w_src_in, irrig_in, w_av_in, p_crop_in, p_hist_in, fert_in, pest_in,
    ]

    # Load dataset: one atomic call returns status + all dropdown updates
    load_btn.click(
        engine.load_dataset,
        inputs=[file_upload],
        outputs=[init_status, *ALL_DROPDOWNS],
    )

    # Cascading updates
    dist_in.change(
        lambda x: gr.update(
            choices=engine.BLOCKS_BY_DISTRICT.get(x, []),
            value=(engine.BLOCKS_BY_DISTRICT.get(x, [None])[0]),
        ),
        inputs=dist_in, outputs=block_in,
    )

    def on_crop_change(c):
        info = engine.CROP_INFO.get(c, {})
        stages    = info.get("stages", [])
        varieties = info.get("varieties", [])
        return (
            gr.update(choices=stages,    value=stages[0]    if stages    else None),
            gr.update(choices=varieties, value=varieties[0] if varieties else None),
        )

    crop_in.change(on_crop_change, inputs=crop_in, outputs=[stage_in, var_in])

    btn.click(
        master_process,
        inputs=[
            dist_in, block_in, vill_in, crop_in, stage_in, var_in, sow_in,
            soil_in, w_src_in, irrig_in, w_av_in, f_cond_in, drain_in, w_log_in,
            p_crop_in, p_hist_in, fert_in, pest_in,
        ],
        outputs=[out_plot1, out_plot2, out_html],
    )

print("\n✅ Launching… A public URL will appear below. Click it to open the app.\n")
demo.launch(share=True, debug=False)
```
